```python
import pandas as pd
import numpy as np
from scipy.sparse import csc_matrix
import matplotlib.pyplot as plt
from IPython.display import clear_output

from tqdm import tqdm_notebook as tqdm
```

```python
df = pd.read_csv('graph.csv')
```

```python
df.head()
```

```python
df
```

```python
rows = df.page_id_to.values
cols = df.page_id_from.values

counts = df.groupby('page_id_from').count()
counts = counts.merge(df, on='page_id_from', suffixes=('', '_'))
counts = counts.page_id_to.values

```

```python
rows = df.page_id_to_labeled.values
cols = df.page_id_from_labeled.values

counts = df.groupby('page_id_from_labeled').count()
counts = counts.merge(df, on='page_id_from_labeled', suffixes=('', '_'))
counts = counts.page_id_to_labeled.values
```

```python
N = max(max(rows), max(cols))+1

beta = 0.8
max_iter = 25
```

```python
groups = np.split(df.values[:, 1], 
                  np.cumsum(np.unique(df.values[:, 0], 
                                      return_counts=True)[1])[:-1])
```

```python
n_ticks = 5
ticks = [x for x in np.arange(0, max_iter+1, max_iter/n_ticks)]    
```

```python
M = csc_matrix((1./counts, (rows, cols)))
v = np.full(N, 1/N)

history = []
for n in tqdm(range(max_iter)):

    # power iteration
    v_new = np.full(N, (1.-beta)/N)
    for i, dest in enumerate(tqdm(groups)):    
        v_new[dest] += beta*v[i]/len(dest)
   
    # compute loss and append to history
    err = v_new-v
    mse = np.mean(err**2)
    history.append([n, mse])
    
    # visualize history of losses 
    clear_output(wait=True)    
    plt.figure(figsize=(12,8))
    plt.plot(np.array(history)[:,0], np.array(history)[:,1])
    plt.xticks(ticks)
    plt.show()
    
    v = v_new
```
